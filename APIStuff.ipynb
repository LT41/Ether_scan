```python
from requests import get
from web3 import Web3
from matplotlib import pyplot as plt
from datetime import datetime
import os
from dotenv import load_dotenv


new_list = []
load_dotenv()

ETHERSCAN_API_KEY = os.getenv('ETHERSCAN_API_KEY')
BASE_URL = "https://api.etherscan.io/api"
address = '0xA3854a218E7faEA99a4ADbEe381bf2AC14F3514f'
ETHER_VALUE = 10 ** 18

#functin to quickly make the different api urls needed
def make_api_URL(module, action, address, **kwargs):
    url = BASE_URL + f"?module={module}&action={action}&address={address}&apikey={ETHERSCAN_API_KEY}"
    
    for key, value in kwargs.items():
        url += f"&{key}={value}"
        
    return url
#returns the ether amount in the specified account 
def get_account_balance(address):
    url = make_api_URL("account", "balance" , address, tag="latest", x="2")
    response = get(url)
    data = response.json()
    
    value = (int(data['result']) / ETHER_VALUE)
    return value
#returns the transactions in the account starting from block 0 and going to the end of the chian. Ether scan self imposes limits to this

# I added a function to it that returns the usd price of the gas spent based on the current price of ethereum.
def get_transactions(address):
    get_transactions_url = make_api_URL('account', 'txlist', address, startblock=0, endblock=99999999, page=1, offset=10000, sort='ascending')
    response = get(get_transactions_url)
    data = response.json()['result']
    
    get_price_url = make_api_URL('stats', 'ethprice', ETHERSCAN_API_KEY)
    response = get(get_price_url)
    data2 = response.json()['result']
    etherprice = float(data2.get('ethusd'))
   
    for tx in data:
        to =tx['to']
        from_adr = tx['from']
        value = int(tx['value']) / ETHER_VALUE
        gas = int(tx['gasUsed']) * int(tx['gasPrice']) / ETHER_VALUE
        gas_usd = (float(tx['gasUsed']) * float(tx['gasPrice']) / ETHER_VALUE) * etherprice
        time = datetime.fromtimestamp(int(tx['timeStamp']))
        new_list.append(gas_usd)
        
        
        print('-----------------------------')
        print('To', to)
        print('From', from_adr)
        print('Value', value)
        print('Gas', gas)
        print('Time', time)
        print(f'USD cost of gas ${gas_usd}')
    print('&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&')
    print(new_list)
    total_usd = sum(new_list)
    print(f'Total spent on gas is ${total_usd}')
        

    
    
    

get_transactions(address)

#adds a string to seperate the transactions
print('******************************************************************')

get_eth_price(address)

```
